```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go
import seaborn as sns
#sns.set_style("dark")
#sns.set_palette("RdBu")
#sns.set_context("notebook", font_scale = 1.5, 
#                rc = { "figure.figsize" : (15, 5), "axes.titlesize" : 18 })
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('howpop_train.csv')

df['published'] = pd.to_datetime(df['published']).dt.date

df['published'] = df['published'].apply(lambda x: datetime.datetime.strftime(x, '%Y-%m'))

tab = df.groupby(['published']).count()

trace = go.Bar(
    x = tab.index,
    y = tab['url']
)
layout = go.Layout(
    title='barchart',
)

fig = go.Figure(data = [trace], layout = layout)
iplot(fig)
```

```python
df = pd.read_csv('howpop_train.csv')

df['published'] = pd.to_datetime(df['published']).dt.date
df['published'] = df['published'].apply(lambda x: datetime.datetime.strftime(x, '%Y-%m-%d'))

df = df[(df['published'] >= '2015-03-01') & (df['published'] <= '2015-03-31')] 
```

```python
df = df.groupby(['domain', 'published']).count()
```

```python


trace0 = go.Scatter(
   # x = df.index,
    y = df.loc['geektimes.ru']['url']
)

trace1 = go.Scatter(
   # x = df.index,
    y = df.loc['habrahabr.ru']['url']
)


# определяем массив данных и задаем title графика в layout
data = [trace0, trace1]
layout = {'title': 'Statistics of geektimes and habr'}

# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
    
```
